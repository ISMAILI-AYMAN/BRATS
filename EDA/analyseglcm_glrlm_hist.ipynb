```python
import pandas as pd
import numpy as np
from scipy import stats

# Assuming you have your data in a DataFrame called 'df'
# with features in columns and tumor grade in a column named 'grade'

# --- Correlation Analysis ---

# 1. Separate data into bas-grade and haut-grade groups:
bas_grade_data = df[df['grade'] == 'bas-grade']  # Replace 'bas-grade' with your actual grade value
haut_grade_data = df[df['grade'] == 'haut-grade']  # Replace 'haut-grade' with your actual grade value

# 2. Select features for analysis:
features = df.columns[df.columns != 'grade']  # Exclude the 'grade' column

# 3. Perform correlation analysis for each feature:
correlation_results = {}
for feature in features:
    # Calculate correlation using Point Biserial Correlation (for continuous feature and binary grade):
    correlation, p_value = stats.pointbiserialr(df['grade'], df[feature])
    correlation_results[feature] = {'correlation': correlation, 'p_value': p_value}

# 4. Identify significant correlations:
significant_features = []
for feature, result in correlation_results.items():
    if result['p_value'] < 0.05:  # Set your significance level (e.g., 0.05)
        significant_features.append((feature, result['correlation'], result['p_value']))

# 5. Print results:
print("Significant Correlations with Tumor Grade:")
for feature, correlation, p_value in significant_features:
    print(f"Feature: {feature}, Correlation: {correlation:.3f}, P-value: {p_value:.3f}")
```
